```python
#%% 2

n1=int(input('ingrese el primer numero:')) #se solicitan los dos numeros enteros
n2=int(input('ingrese el segundo numero:'))

m=abs(n1-n2)     #m es la diferencia entre los dos numeros y mediante el comando abs se logra que esa cantidad siempre sea positiva por el valor absoluto

for i in range(2,m): #se crea un ciclo for que tenga rango desde 2 hasta la diferencia por ue necesitamos dividir el numero entre todos los numeros antes de m
    if m%i != 0:  # se divide la diferencia entra cada numero antes de la misma y si el residuo es 0 entonces no sera un numero primo     
        print('es un numero primo')
    else:
        print('no es un numero primo')
    break
```

```python
#%% 3                                  revisar 
def multiple(valor, multiple):
    """
    Funcion para calcular si el numero es multiplo
    utilizando el modulo de la division
    """
    resto = valor % multiple   
    if resto == 0:      #si el resto entre los dos anteriores es 0 significa que uno es multiplo del otro
        return True
    else:
        return False
 
# lista que contendra los valores multiples
multiples_3=[]  
multiples_4=[]
 
# bucle del 1 al 100
for i in range(1,101):
 
    if multiple(i,3):    
        multiples_3.append(i)  # mediante append agregamos cada numero que este en el rango solicitado y cumpla la condicion a la lista de multiplos de 3 
 
    if multiple(i,4):
        multiples_4.append(i)
#se imprimen los resultados
print ("Los multiples de 3 son:", multiples_3)   
print( "")
print( "Los multiples de 4 son:", multiples_4)
```

```python
#%% 5
import keyword 
def leer_frase():
    global txt  # se almacena la variable txt como global
    txt = (input('ingrese un texto:'))     
    
def contar_letras():
    #se cran todos los contadores necesarios para ejecutar el programa
    conta = 0      
    conta_tildes=0
    conta_vocal=0
    conta_mayus=0
    conta_digit=0
    conta_space=0
    conta_reservadas=0
    for i in txt:
        if (i.isalpha()): #con is alpha nos dice si lo que esta en i es alfabetico
            conta += 1
        if i=='a' or i=='e' or i=='i' or i=='o' or i=='u': #aqui le decimos al programa que cada que encuentre una de estas letras(vocales) aunmente el contador
            conta_vocal +=1
        if (i.isupper()): #con isuper sabemos si hay mayusculas
            conta_mayus +=1
            
    for j in txt:
        if j=='á'or j=='é' or j=='í' or j=='ú' or j=='ó':  #aqui sabremos cuantas tildes hay y las coloco solo en las vocales por que siempre las tildes van en las vocales
            conta_tildes +=1
    
    for x in txt:
        if (x.isdigit()): #xon is digit sabremos si lo ingresado contiene digitos 
            conta_digit +=1
    
    for y in txt:
        if (y.isspace()): #aqui sabremos cuantos espacios hay 
            conta_space +=1
            
    for z in txt:
        if keyword.iskeyword(z): #este comando nos permite conocer cuantas palabras reservadas hay en el texto
            conta_reservadas +=1
    
            
    print('La cantidad de letras es:',conta)
    print('la cantidad de vocales es:',conta_vocal)
    print('la cantidad de mayusculas es:',conta_mayus)
    print('la cantidad de tildes es:',conta_tildes)
    print('la cantidad de digitos es:',conta_digit)
    print('la cantidad de espacios es:',conta_space)    
    print('la cantidad de palabras reservadas es:',conta_reservadas)
    
    
```

```python
#%%13
def biciestos(año_1,año_2):  
    "se definen que la funcion depende de año_1 y año_2"
    try:
        
        año_1=int(input('igrese el año inicial:'))  # en estas dos lineas se solicita la informacion necesaria para el programa
        año_2=int(input('igrese el año final:'))
        conta=0        #se crea el contador que sera el responsable de almacenar cuantos años hay
        for i in range(año_1,año_2+1): #mediante este ciclo le decimos al programa que la funcion i toma valores entre año_1 y año_2 incluyendo el 2 gracias al +1
            if i%4==0 :      #debido a que todos los años biciestos son multiplo de 4 entoncs se utiliza el comando de divicion entera
                conta += 1   #en eta linea cada vez que el if sea "True" es decir cada vez que este se cumpla el cintador aumentara 1 año mas
        
        print('los años biciestos son.',conta) #se imprime el contador
    except ValueError:          #en caso de que el usuario ingre un valor que no sea un año le dira que no se puede hacer
        print('no puede ingresar letras')
biciestos(año_1,año_2)  
```

```python
#%%14

import random

def sumar():
    "se define la funcion que sumara los numeros positivos"
    print('escriba el resultado de las siguientes operaciones:')
    #se crea el contador
    correcta=0
    #se crea un ciclo que funcionara hasta que el usuario acierte 5 sumas
    while correcta < 5:
        #se generan numeros aleatorios en un rango entre 1 y 100
        a= random.randrange(1,101) 
        b= random.randrange(1,101)
        respuesta= int(input(f'{a} + {b} = ')) 
        #se cran los condicionales para la respuesta que de el usuario 
        if respuesta==a + b: 
            correcta += 1
            print('respuesta correcta')
        else:
            print('respuesta incorrecta')
            break
        print()
    print('fin')
if __name__=='__main__':
    sumar()
```
